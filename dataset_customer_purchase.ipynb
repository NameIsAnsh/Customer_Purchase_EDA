## Grouping the customers according to the age group

```python
age_groups = [18, 30, 45, 60, 80]
age_labels = ['18-30', '30-45', '45-60', '60-80']

df['Age_Group'] = pd.cut(df['Age'], bins=age_groups, labels=age_labels)

grouped_data = df.groupby('Age_Group')[['Spending_Score', 'Income']].mean()

print(grouped_data)
```

### Spending Score across the Age Group 

```python

grouped_data['Spending_Score'].plot(kind='bar', color='lightblue')
plt.title('Average Spending Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending Score')
plt.xticks(rotation=45)
plt.show()
```

```python
high_value= df[(df['Spending_Score'] >df['Spending_Score'].quantile(0.75)) & (df['Income']>df['Income'].quantile(0.75))]
high_value
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Income', y='Spending_Score',hue='Age_Group', data=high_value, palette='cool')
plt.title('Spending Score vs. Income by Age Group for High Value Customers')
plt.show()
```
